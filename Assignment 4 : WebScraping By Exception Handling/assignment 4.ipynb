```python
Question 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
name=[]

element=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in element:
        name.append(i.text)
name
```

```python
artist=[]
date=[]
view=[]

element2=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for i in element2:
    artist.append(i.text)
    
element3=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in element3:
    view.append(i.text)
    
element4=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in element4:
    date.append(i.text)
```

```python
rank=[]
start=1
end=31
for i in range(start,end):
    rank.append(i)
```

```python
import pandas as pd
df=pd.DataFrame({'Rank':rank,'Video Name':name,'Artist Name':artist,'Total Views':view,'Upload Date':date})
df
```

```python
driver.close()
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.imdb.com/chart/tvmeter/?ref_=nv_tvv_mptv/')
```

```python
name=[]
episode=[]
rating=[]


element1=driver.find_elements(By.XPATH,'//li[@class="ipc-metadata-list-summary-item sc-1364e729-0 caNpAE cli-parent"]/div[2]/div/div/div[2]')
for i in element1:
    name.append(i.text)
    
element2=driver.find_elements(By.XPATH,'//li[@class="ipc-metadata-list-summary-item sc-1364e729-0 caNpAE cli-parent"]/div[2]/div/div/div[3]/span[2]')
for i in element2:
    episode.append(i.text)
    
element3=driver.find_elements(By.XPATH,'//div[@class="ipc-metadata-list-summary-item__c"]/div/div/span/div/span')
for i in element3:
    rating.append(i.text)
```

```python
url=[]
for i in driver.find_elements(By.XPATH,'//div[@class="ipc-metadata-list-summary-item__c"]/div/div/div[2]/a'):
    url.append(i.get_attribute('href'))
print(len(url))
```

```python
genre=[]
rating=[]
vote=[]
year=[]

for i in url:
    driver.get(i)
    time.sleep(3)
    
    try:
        element1=driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]')
        genre.append(element1.text)
    except NoSuchElementException:
        genre.append('-')
        
    try:
        element2=driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[1]/div/div[1]/a/span/div/div[2]/div[1]/span[1]')
        rating.append(element2.text)
    except NoSuchElementException:
        rating.append('-')
        
    try:
        element3=driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[1]/div/div[1]/a/span/div/div[2]/div[3]')
        vote.append(element3.text)
    except NoSuchElementException:
        vote.append('-')
        
    try:
        element4=driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]/a')
        year.append(element4.text)
    except NoSuchElementException:
        year.append('-')
```

```python
import pandas as pd
df=pd.DataFrame({'Name':name,'Year span':year,'Genre':genre,'Total Episodes':episode,'Ratings':rating,'Votes':vote})
df
```

```python
8. Details of Datasets from UCI machine learning repositories. 
Url = https://archive.ics.uci.edu/ You 
have to find the following details: 
A) Dataset name 
B) Data type 
C) Task 
D) Attribute type 
E) No of instances 
F) No of attribute G) Year 
 Note: - from the home page you have to go to the Show All Dataset page through code.
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
click=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
click.click()
```

```python
name=[]
types=[]
instance=[]
attribute=[]

element1=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in element1:
    name.append(i.text)
    
element2=driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/div/div[2]')
for i in element2:
    types.append(i.text)
    
element3=driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/div/div[3]')
for i in element3:
    instance.append(i.text)
    
element4=driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/div/div[4]')
for i in element4:
    attribute.append(i.text)
```

```python
url=[]
for i in driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]'):
    url.append(i.get_attribute('href'))
url
```

```python
task=[]
attribute_type=[]
year=[]

for i in url:
    driver.get(i)
    time.sleep(3)
    
    try:
        element5=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p')
        task.append(element5.text)
    except NoSuchElementException:
        task.append('-')
        
    try:
        element6=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p')
        attribute_type.append(element6.text)
    except NoSuchElementException:
        attribute_type.append('-')
        
    try:
        element7=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/h2')
        year.append(element7.text)
    except NoSuchElementException:
        year.append('-')
```

```python
instance=[]
attribute=[]

for i in url:
    driver.get(i)
    time.sleep(3)
    
    try:
        element8=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p')
        instance.append(element8.text)
    except NoSuchElementException:
        instance.append('-')
        
    try:
        element9=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p')
        attribute.append(element9.text)
    except NoSuchElementException:
        attribute.append('-')
```
