Question 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search 
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then 
scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
product_url=[]

start=0
end=3
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        product_url.append(i.get_attribute('href'))
        
    next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[29]/div/div/span/a[3]')
    
product_url
```

```python
len(product_url)
```

```python
brand_name=[]
product_name=[]
total_price=[]
rating=[]

for i in product_url:
    driver.get(i)
    time.sleep(2)
    
    #Extracting Brand Name
    try:
        brand = driver.find_element(By.XPATH,'/html/body/div[2]/div/div[5]/div[3]/div[4]/div[46]/div/table/tbody/tr[1]/td[2]/span')
        brand_name.append(brand.text)
    except NoSuchElementException:
        brand_name.append('-')
        
    #Extracting Product name
    try:
        product = driver.find_element(By.XPATH,'/html/body/div[2]/div/div[5]/div[3]/div[4]/div[1]/div/h1/span')
        product_name.append(product.text)
    except NoSuchElementException:
        product_name.append('-')
        
    #Extracting total price
    try:
        price = driver.find_element(By.XPATH,'/html/body/div[2]/div/div[5]/div[3]/div[4]/div[13]/div/div/div[4]/div[1]/span[3]/span[2]/span[2]')
        total_price.append(price.text)
    except NoSuchElementException:
        total_price.append('-')
        
    #Extracting Ratings
    try:
        rate = driver.find_element(By.XPATH,'/html/body/div[9]/div/div[1]/div/div/div/div/div[1]/span')
        rating.append(rate.text)
    except NoSuchElementException:
        rating.append('-')
```

```python
import pandas as pd
df=pd.DataFrame({'Brand Name':brand_name,'Product Info':product_name,'Product Price':total_price,'Product_url':product_url})
df
```

Question 9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in 
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall 
reviews, privates from price, dorms from price, facilities and property description. 

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.hostelworld.com/')
```

```python
element=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input')
element.send_keys('London')
```

```python
element1=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[5]/button[1]')
element1.click()
```

```python
url=[]

for i in driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]'):
    url.append(i.get_attribute('href'))
url
```

```python
hostal_name=[]
distance=[]
rating=[]
total_review=[]
overall_review=[]


#Extracting Hostal Name
try:
    for i in driver.find_elements(By.XPATH,'//div[@class="property-name"]/span'):
        hostal_name.append(i.text)
except NoSuchElementException:
    hostal_name.append('-')
    
#Extracting Distance
try:
    for i in driver.find_elements(By.XPATH,'//span[@class="distance-description"]'):
        distance.append(i.text.replace('-',''))
except NoSuchElementException:
    distance.append('-')
    
#Extracting rating
try:
    for i in driver.find_elements(By.XPATH,'//span[@class="number"]'):
        rating.append(i.text)
except NoSuchElementException:
    rating.append('-')
    
#Extracting Total Review
try:
    for i in driver.find_elements(By.XPATH,'//span[@class="left-margin"]'):
        total_review.append(i.text)
except NoSuchElementException:
    total_review.append('-')
```

```python
overall_review=[]
private_price=[]
dorms_price=[]
facility=[]
description=[]

for i in url:
    driver.get(i)
    time.sleep(5)
    
    #Extracting Overall Review
    try:
        element1=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[8]/div/div[1]/div[1]/div[2]/div[1]/span')
        overall_review.append(element1.text)
    except NoSuchElementException:
        overall_review.append('-')
        
    #Extracting Private Price
    try:
        element2=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[10]/div/div/section/div/div[1]/div/div[1]/div[2]/div[1]/div[2]/div[2]/div/div[1]/div[2]/div[1]')
        private_price.append(element2.text)
    except NoSuchElementException:
        private_price.append('-')
        
    #Extracting Dorms Price
    try:
        element3=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[10]/div/div/section/div/div[1]/div/div[2]/div[2]/div/div[2]/div[2]/div/div[1]/div[2]/div[1]')
        dorms_price.append(element3.text)
    except NoSuchElementException:
        dorms_price.append('-')
        
    #Extracting Facility
    try:
        element4=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[12]/div/ul/li[1]/ul')
        facility.append(element4.text)
    except NoSuchElementException:
        facility.append('-')
        
    #Extracting Description
    try:
        element5=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[8]/div/div[2]/div[3]/div/div[2]')
        description.append(element5.text)
    except NoSuchElementException:
        description.append('-')
```

```python
import pandas as pd

df=pd.DataFrame({'Hostal Names':hostal_name,'Distance From City Centre':distance,'Ratings':rating,'Total Ratings':total_review,'Overall_Review':overall_review,'Privates Price':private_price,'Dorms Price':dorms_price,'Facilities':facility,'Description':description})
```
